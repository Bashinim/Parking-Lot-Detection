```python
# Importing libraries
from PIL import Image
import PIL
PIL.Image.MAX_IMAGE_PIXELS = 400000000
import cv2
import numpy as np
```

```python
#Converting relevant tif images to PNG images

im = Image.open('E:/Berlin-DOP_2020/Processed images/dop10rgbi_391_5821_1_be_2020.tif')
im.convert('RGB').save('E:/Berlin-DOP_2020/Processed images/dop10rgbi_391_5821_1_be_2020.png')
del im

im = Image.open('E:/Berlin-DOP_2020/Processed images/dop10rgbi_391_5821_1_be_2020_roads.tif')
im.convert('RGB').save('E:/Berlin-DOP_2020/Processed images/dop10rgbi_391_5821_1_be_2020_roads.png')
del im
```

```python
# Converting the Roads raster to a binary mask

# load image with alpha channel
img = cv2.imread('E:/Berlin-DOP_2020/Processed images/dop10rgbi_391_5821_1_be_2020_roads.png', cv2.IMREAD_UNCHANGED)

# extract alpha channel
alpha = img[:, :]

# threshold alpha channel
alpha = cv2.threshold(alpha, 0, 255, cv2.THRESH_BINARY)[1]

#Inverting the mask
mask = cv2.bitwise_not(alpha)

# save output
cv2.imwrite('E:/Berlin-DOP_2020/Processed images/dop10rgbi_391_5821_1_be_2020_roads_mask.png', mask)
```

```python
#Applying the mask to the image

img = cv2.imread('E:/Berlin-DOP_2020/Processed images/dop10rgbi_391_5821_1_be_2020_roads.png')
mask = cv2.imread('E:/Berlin-DOP_2020/Processed images/dop10rgbi_391_5821_1_be_2020_roads_mask.png',0)
res = cv2.bitwise_and(img,img,mask = mask)

# Saving the putput
cv2.imwrite('E:/Berlin-DOP_2020/Processed images/Maskeddop10rgbi_391_5821_1_be_2020_roads.png', res)
```
